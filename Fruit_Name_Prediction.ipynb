**FRUIT NAME PREDICTION BY KAUMUDI**

```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
```

```python
# Load your dataset
df = pd.read_csv("fruit_classification_dataset.csv")
```

```python
df.info()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
# Label Encoding
df["fruit_name"]=label_encoder.fit_transform(df["fruit_name"])
df["color"]=label_encoder.fit_transform(df["color"])
df["season"]=label_encoder.fit_transform(df["season"])
df["shape"]=label_encoder.fit_transform(df["shape"])
df["taste"]=label_encoder.fit_transform(df["taste"])
df["skin_type"]=label_encoder.fit_transform(df["skin_type"])
```

```python
df
```

```python
#Creating x and y

x=df[[ 'color', 'season', 'average_weight_g', 'shape', 'taste',
       'size_cm', 'is_seeded', 'skin_type']]

y=df["fruit_name"]
```

```python
x
```

```python
y
```

```python
# Train Test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
knn_model=KNeighborsClassifier()
```

```python
knn_model.fit(x_train, y_train)
```

```python
# model predicting
y_pred=knn_model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import pickle
with open("fruit_prediction.pkl","wb") as f:
  pickle.dump(knn_model,f)
```
